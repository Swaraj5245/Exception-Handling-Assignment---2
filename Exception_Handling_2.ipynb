### Q1. Explain why we have to use the Exception class while creating a Custom Exception.
**Note: Here Exception class refers to the base class for all the exceptions.**

WE use Exception class for creating a custom Exception because when the custom error raise, it raise as a Exception, Not as only the message.  
For that we inherite Exception Base class in the Custom Class to make that class as a Exception class. that custom Exception class takes argument as a message and that message will be the custom message which we wants to show when the Error raise.  

**Example:**

```python
class CustomException(Exception):
    def __init__(self, message):
        self.message=message
```

### Q2. Write a python program to print Python Exception Hierarchy.

```python
def Exception_Hierarchy(exception_class):
    print(exception_class.__name__)
    
    for subclass in exception_class.__subclasses__():
        Exception_Hierarchy(subclass)

print("Python Exception Hierarchy:")
Exception_Hierarchy(Exception)
```

### Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a subclass of a Exception class, it is used to raise error when there is no module associated with that name. For raising import error we use ModuleNotFoundError in the except block.
ModuleNotFoundError used to raise error when the module not exists with that name.  

**Example:**

```python
try:
    import swaraj
except ModuleNotFoundError as e:
    print(e)
```
